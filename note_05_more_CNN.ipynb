```python
nx = 128
ny = 128
nc = 3
```

```python
model2 = Sequential()
model2.add(Conv2D(6, kernel_size=(5, 5), strides=(2, 2), activation='relu', input_shape=(ny, nx, 1)))
model2.add(Conv2D(12, kernel_size=(5, 5), strides=(2, 2), activation='relu'))
model2.add(Conv2D(24, kernel_size=(5, 5), strides=(2, 2), activation='relu'))
model2.add(Flatten())
model2.add(Dense(64, activation='relu'))
model2.add(Dense(nc, activation='softmax'))
model2.summary()
model2.compile(loss=keras.losses.categorical_crossentropy,
             optimizer='adagrad',
             metrics=['accuracy'])
```
